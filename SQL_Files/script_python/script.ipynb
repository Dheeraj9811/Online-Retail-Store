```python
import random
import names

def random_count():
    return random.randint(1, 10)

def random_total_cost():
    return random.uniform((1, 100), 2)

def random_product_id():
    return random.randint(1, 100)

def generate_cart_data(num_customers):
    sample_data = []
    for i in range(num_customers):
        c_id = i + 1
        ca_id = i + 1
        count = random_count()
        total_cost = random_total_cost()
        sample_data.append(f"({ca_id}, {c_id}, {count}, {total_cost})")
    return ',\n'.join(sample_data)

def generate_cart_product_data(num_customers):
    sample_data = []
    for i in range(num_customers):
        ca_id = i + 1
        cp_id = i + 1
        p_id = random_product_id()
        count = random_count()
        sample_data.append(f"({cp_id}, {ca_id}, {p_id}, {count})")
    return ',\n'.join(sample_data)

def write_to_file(data, filename):
    with open(filename, 'w') as file:
        file.write(data)

num_customers = 100
cart_data = generate_cart_data(num_customers)
insert_query_cart = "INSERT INTO Cart (Ca_id, C_id, COUNT, T_cost) VALUES \n" + cart_data + ";"
write_to_file(insert_query_cart, 'cart_data.txt')

cart_product_data = generate_cart_product_data(num_customers)
insert_query_cart_product = "INSERT INTO Cart_Products (Cp_id, Ca_id, P_id, COUNT) VALUES \n" + cart_product_data + ";"
write_to_file(insert_query_cart_product, 'cart_product_data.txt')
```

# Order

```python
import random
import names
import datetime

# Generate 100 unique customer IDs
customer_ids = list(range(1, 101))

# Create a list to store the INSERT statements
insert_statements = []

for i in range(1, 101):
    # Generate a random date for the order and shipping date
    order_date = str(datetime.date.today() + datetime.timedelta(days=random.randint(1, 365)))
    ship_date = str(datetime.date.today() + datetime.timedelta(days=random.randint(1, 365)))
    
    # Generate a random total cost for the order
    total_cost = round(random.uniform(10.00, 1000.00), 2)
    
    # Choose a random customer ID from the list of customer IDs
    customer_id = random.choice(customer_ids)
    
    # Remove the chosen customer ID from the list so that it is not chosen again
    customer_ids.remove(customer_id)
    
    # Create the INSERT statement for this order
    insert_statement = f"INSERT INTO Order_ (C_id, O_date, Ship_date, T_cost) VALUES ({customer_id}, '{order_date}', '{ship_date}', {total_cost});"
    
    # Add the INSERT statement to the list of insert statements
    insert_statements.append(insert_statement)

# Write the insert statements to a text file, with each statement on a new line
with open("order_data.txt", "w") as file:
    for statement in insert_statements:
        file.write(statement + "\n")
```

# Supplier


```python
import names
import random

# generate 100 random Suppliers data
num_of_entries = 100

# create a list to store the Suppliers data
suppliers = []

for i in range(num_of_entries):
    # generate a random name using the `names` library
    supplier_name = names.get_full_name()

    # generate a random product ID
    product_id = random.randint(1, 100)

    # generate a random quantity
    quantity = random.randint(1, 1000)

    # generate a random phone number
    phone_number = "".join([str(random.randint(0, 9)) for i in range(10)])

    # generate a random email
    email = supplier_name.lower().replace(" ", ".") + "@email.com"

    # add the Supplier data to the list
    suppliers.append([supplier_name, product_id, quantity, phone_number, email])

# write the Suppliers data to a text file
with open("suppliers.txt", "w") as f:
    for supplier in suppliers:
        f.write("INSERT INTO Suppliers (Sup_name, P_id, Quantity, Phn_no, Email_id) VALUES ('{}', {}, {}, '{}', '{}');\n".format(*supplier))
```
